## Read the image 

```python
#load the image 
input_image = cv2.imread('lena.png')
print(type(input_image))
```

- this is used in Image processing because of the contrast
- it returns numpy array 
    - it's a 3D numpy array 
        - x and y are the axis 
        - z is the height (which is 3 for the 3 channels
- images in general are rgb 
    - matplot lib when it reads and draws the image it reads it rgb 
    - but OpenCv when it reads the image, it reads bgr (that means the red will be read as blue and the blue will be read as red)
        - so, we need to swap the first and last channels

```python
# using matplot lib 
plt.imshow(input_image)
```

- we have the above result because as we said, opencv reads bgr

## plotting actual gray image 

```python
gray_lena = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
gray_lena.shape
```

the function wheights each channel with a value then add them together 

```python
plt.imshow(gray_lena, cmap='gray')
```

### playing with it a little bit 

```python
gray_lena
```

```python
gray_lena[0,0]
```

```python
lightened_lena = gray_lena
lightened_lena
```

```python
for idx, pixel in np.ndenumerate(gray_lena):
    if pixel < 200:
        lightened_lena[idx] += 50
```

## Fix the awkward error BGR to RGB

```python
fixed_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_image)
```

### HSV Colorspace

```python
hsv_img = cv2.cvtColor(fixed_image, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
```

- we see a wierd result, because the first  channel red will be treated as hue, green will be considered saturation and blue will be value 

```python
hue, saturation, value = cv2.split(hsv_img)
```

```python
plt.imshow(hue)
```

```python
plt.imshow(saturation)
```

```python
plt.imshow(value)
```

## Cropping the image 

```python
plt.imshow(fixed_image)
```

we see that the face range is from 200 to 400 vertically and from 200 to 350 horizontally 

```python
plt.imshow(fixed_image[200:400,200:350]) #the index is the usual array[x,y] where x is the vertical and y is the horizontal axis
```

## Flip the image 

### vertically 

```python
flipped_v = cv2.flip(fixed_image,0) # 0 is the axis, vertically is axis 0 -beacause index 0 is the index that axis the vertical part in the 2d array-
plt.imshow(flipped_v)
```

### horiontally 

```python
flipped_v = cv2.flip(fixed_image,1) # 1 is the axis, horizontally is axis 1
plt.imshow(flipped_v)
```

## Transpose the image 

```python
trans = cv2.transpose(fixed_image)
plt.imshow(trans)
```

## Noise reduction (blurring)

```python
plt.imshow(fixed_image)
```

```python
k = 3 #kernel size, better to be odd 
blurred_img = cv2.GaussianBlur(fixed_image, (k,k), -1) #-1 is for sigmax
plt.imshow(blurred_img)
```

- the kernel size is better to be odd
    - mathimatically speaking it is true 
    - also practically is true 
    - but finds the mean and median indirectly

```python
k = 5 #kernel size, better to be odd 
blurred_img = cv2.GaussianBlur(fixed_image, (k,k), -1) #-1 is for sigmax
plt.imshow(blurred_img)
```

```python
k = 31 #kernel size, better to be odd 
blurred_img = cv2.GaussianBlur(fixed_image, (k,k), -1) #-1 is for sigmax
plt.imshow(blurred_img)
```
