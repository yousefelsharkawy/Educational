```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random = np.random.randint(0,256,(240,320,3),dtype='uint8')
```

```python
plt.imshow(random)
```

## save as JPEG

```python
plt.imsave('random.jpg',random)
```

```python
random_jpg = plt.imread('random.jpg')
plt.imshow(random_jpg)
```

```python
random_jpg.shape
```

```python
[random_jpg.min(),random_jpg.max()]
```

```python
# get the number of different pixels 
(random_jpg != random).sum()
```

```python
# get the mean of different pixels 
abs(random_jpg-random).mean()
```

we notice that most of the pixels are different
- this is because it is a random image 
- and jpeg procedure works the worst with it 
    - as it works with dct (discrete cosine transform)
    - since the image is random, dct wont work with it 
    - because it works with normal images (where the high frequencies are concentrated in the middle
    - but for random image, the FT will be 

## save as gif (lossy)

```python
plt.imsave('random.gif',random)
```

```python
random_gif = plt.imread('random.gif')
plt.imshow(random_gif)
```

```python
print(random_gif.shape)
print([random_gif.min(),random_gif.max()])
# get the number of different pixels 
print((random_gif != random).sum())# get the mean of different pixels 
print(abs(random_gif-random).mean())
```
