```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
from skimage.io import imread,imshow,imsave
```

```python
im = imread('walking.jpg')
plt.figure()
plt.gray()
plt.imshow(im)
plt.show()
```

```python
from scipy.signal import convolve2d
from skimage.filters import gaussian

gy = np.array([[-1.,-2.,-1.],[0.,0.,0.],[1.,2.,1.]])
gx = gy.T # transpose operator

print(gy)
print(gx)
```

```python
Gy = convolve2d(im, gy)
Gx = convolve2d(im, gx)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(np.abs(Gy))
plt.subplot(1,2,2)
plt.imshow(np.abs(Gx))
plt.show()
```

```python
s = np.sqrt(Gy**2+Gx**2)
d = np.arctan2(Gy,Gx)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(s)
plt.subplot(1,2,2)
plt.imshow(d, cmap=plt.cm.viridis)
plt.colorbar()
plt.show()
```
