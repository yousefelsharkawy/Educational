```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook

from skimage.io import imread,imshow,imsave

im = imread('walking.jpg')
print(im.shape)

plt.figure()
imshow(im)
plt.show()
```

```python
from scipy.signal import convolve2d

# Mean kernel
ksize = 15
kernel = np.ones((ksize,ksize))
kernel /= kernel.sum()

im_out = convolve2d(im, kernel)

plt.figure()
plt.gray()
plt.subplot(1,2,1)
plt.imshow(im[::5,::5])
plt.subplot(1,2,2)
plt.imshow(im_out[::5,::5])
plt.show()
```

```python
from scipy.signal import convolve2d
from skimage.filters import gaussian

# Mean vs Gaussian
im_out2 = gaussian(im, 3)

plt.figure()
plt.gray()
plt.subplot(1,2,1)
plt.imshow(im_out[::5,::5])
plt.subplot(1,2,2)
plt.imshow(im_out2[::5,::5])
plt.show()
```
