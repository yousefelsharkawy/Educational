```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
from skimage.io import imread,imshow

im = imread('etretat.jpg')
plt.figure()
imshow(im)
plt.show()
```

```python
# Channels & histograms
print(im.shape, im.dtype)

plt.figure()
plt.subplot(1,3,1)
plt.imshow(im[:,:,0], plt.cm.gray)
plt.subplot(1,3,2)
plt.imshow(im[:,:,1], plt.cm.gray)
plt.subplot(1,3,3)
plt.imshow(im[:,:,2], plt.cm.gray)
plt.show()
```

```python
plt.figure()
plt.subplot(3,1,1)
plt.hist(im[:,:,0].flatten(), bins=np.arange(257))
plt.subplot(3,1,2)
plt.hist(im[:,:,1].flatten(), bins=np.arange(257))
plt.subplot(3,1,3)
plt.hist(im[:,:,2].flatten(), bins=np.arange(257))
plt.show()
```

```python
from skimage.color import rgb2hsv

hsv = rgb2hsv(im)
print(hsv.shape, hsv.dtype)
```

```python
# Channels & histograms
plt.figure()
plt.subplot(1,3,1)
plt.imshow(hsv[:,:,0], plt.cm.hsv)
plt.subplot(1,3,2)
plt.imshow(hsv[:,:,1], plt.cm.gray)
plt.subplot(1,3,3)
plt.imshow(hsv[:,:,2], plt.cm.gray)
plt.show()
```

```python
plt.figure()
plt.subplot(3,1,1)
plt.hist(hsv[:,:,0].flatten(), bins=np.arange(101)/100)
plt.subplot(3,1,2)
plt.hist(hsv[:,:,1].flatten(), bins=np.arange(101)/100)
plt.subplot(3,1,3)
plt.hist(hsv[:,:,2].flatten(), bins=np.arange(101)/100)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

def scatter_3d(coords, colors):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(coords[:,0], coords[:,1], coords[:,2], s=2, c=colors)
    plt.show()
```

```python
pixels = im.reshape((im.shape[0]*im.shape[1],3))[::20,:]

scatter_3d(pixels, pixels/255)
```

```python
pixelshsv = hsv.reshape((im.shape[0]*im.shape[1],3))[::20,:]

scatter_3d(pixelshsv, pixels/255)
```
