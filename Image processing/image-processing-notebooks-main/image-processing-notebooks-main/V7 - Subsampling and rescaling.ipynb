```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
from skimage.io import imread,imshow
```

```python
im = imread('camera.jpg')
print(im.shape)
plt.figure()
imshow(im)
plt.show()
```

```python
def subsample(im, factor):
    return im[::factor,::factor]

im_s = subsample(im, 2)
print(im_s.shape)
print(im_s.dtype)
plt.figure()
imshow(im_s)
plt.show()
```
