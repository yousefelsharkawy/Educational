```python
import numpy as np
```

```python
a = np.zeros((3,2))
print("a")
print(a)
b = np.ones((3,2))
print("b")
print(b)
c = np.random.random((3,2))
print("c")
print(c)
d = np.array([[1,2],[3,4],[5,6]])
print("d")
print(d)
```

```python
a = np.array([[5,2],[3,4],[1,0]])
b = np.array([[8,-3],[-1,12],[40,20]])
print("a")
print(a)
print("b")
print(b)
print("a+b -> per-element operation")
print(a+b)
print("a+5 -> scalar operation distributed to all elements")
print(a+5)
```

```python
from skimage.io import imread,imshow,imsave
%matplotlib notebook
from matplotlib import pyplot as plt

im = imread('walking.jpg')
plt.figure()
imshow(im)
plt.figure()
imshow(im+50)
plt.show()
```

```python
print(im[500,200])
window = im[355:560,140:260]

plt.figure()
imshow(window)
plt.show()
```

```python
im3 = im.copy()
im3[355:560,265:385] = window
plt.figure()
imshow(im3)
plt.show()
```

```python
mask = im<30

plt.figure()
imshow(mask)
plt.show()

im4 = im.copy()
im4[mask] = 255

plt.figure()
imshow(im4)
plt.show()
```

```python
print(im)
print("Indexing with scalar coordinates:")
print(im[5,5], im[5,5].shape)
print("Indexing with scalar range:")
print(im[5:10,5:10], im[5:10,5:10].shape)
print("Indexing with boolean mask:")
print(im[mask], im[mask].shape)
```
