```python
import matplotlib.pyplot as plt
import seaborn as sns

# load the logs file
with open('logs.txt', 'r') as f:
    # read the lines of the file
    logs = [line.strip() for line in f.readlines()]

train_loss = []
val_loss = []
val_indices = []
hella = []
hella_indices = []

for log in logs:
    iteration, type, value = log.split(' ')
    if type == 'train':
        train_loss.append(float(value))
    elif type == 'val':
        val_loss.append(float(value))
        val_indices.append(int(iteration))
    elif type == 'hella':
        hella.append(float(value))
        hella_indices.append(int(iteration))
```

```python
# Set the Seaborn style
sns.set_theme(palette='colorblind')
```

```python
# plot the hella accuracy
plt.plot(hella_indices, hella, label='Our GPT-2 (124M) checkpoint')
plt.axhline(y=0.2955, color='r', linestyle='--', label='OpenAI GPT-2 (124M) checkpoint')
plt.axhline(y=0.337, color='g', linestyle='--', label='OpenAI GPT-3 (124M) checkpoint')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.legend()
plt.title('HellaSwag Accuracy')
```

```python
# plot the train and validation loss
plt.figure()
plt.plot(train_loss, label='train')
plt.plot(val_indices, val_loss, label='validation')
plt.title('Loss')
plt.legend()
```
