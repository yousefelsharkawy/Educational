- spearman correlation means the correlation between the ranks of two variables.
- The spearman correlation is a non-parametric measure of rank correlation (statistical dependence between the rankings of two variables).
- we can get it in pandas using the corr() method with the method parameter set to spearman.
    - `df.corr(method='spearman')`

```python
df.corr(method="spearman")
```
