```python
# import and load data
import pandas as pd
%matplotlib inline
df = pd.read_csv("powerplant_data_edited.csv")
df.info()
df.head()
```

```python
# plot relationship between temperature and electrical output
df.plot(x="ambient_Temprature",y="electrical_energy_output",kind="scatter");
```

- we see that there is a negative correlation between temperature and net electrical output
    - this means that as temperature increases, net electrical output decreases

- Does the humidity variable appear to be normally distributed?
    - we said that in order to see the distribution, we need to plot a histogram

```python
# plot distribution of humidity
df["Relative_Humidity"].plot(kind="hist")
```

- Which variable appears to have the most outliers?

```python
# plot box plots for each variable
df.boxplot(column=['ambient_Temprature','Exhaust _vacuum','Relative_Humidity',"Ambient_Pressure"])
```

- it is hard to plot all of them together as they have different ranges , so we can plot each one of them separately

```python
df["Exhaust _vacuum"].plot(kind="box");
```
